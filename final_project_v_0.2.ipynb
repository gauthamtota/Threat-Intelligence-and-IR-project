```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('Android_Permission.csv')
dataset.head()
```

```python
dataset = dataset.drop(['App','Package','Description','Related apps'],axis=1)
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(dataset['Category'])
dataset['Category'] = encoded_labels
dataset.head()
```

```python
na_count = [(column, dataset[column].isna().sum()) for column in dataset.columns]
na_count
```

Removing NA values

```python
dataset = dataset.dropna(axis=0)
dataset.info()
na_count = dataset.isna().sum().sum()
print(na_count)
```

Removing the rows with all zero values

```python
dataset = dataset.loc[:, (dataset != 0).any(axis=0)]
dataset.describe()
```

Creating X and y for analysis

```python
df = dataset
X = df.drop('Class',axis = 1)
y = df['Class']
X
```

Using SMOTE to balance the dataset

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=23)
X, y = smote.fit_resample(X, y)
X
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlations
correlations = X.corrwith(y)

# Sort the correlations in descending order
correlations = correlations.sort_values(ascending=False)
print(correlations)

# Select top 9 and bottom 9 correlated features
top_features = correlations[:9]
bottom_features = correlations[-9:]

# Combine top and bottom features
selected_correlations = pd.concat([top_features, bottom_features])

# Convert to DataFrame
selected_df = selected_correlations.reset_index()
selected_df.columns = ['Features', 'Correlations']

# Plot the correlations using a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=selected_df, x='Features', y='Correlations')
plt.xticks(rotation=90)
plt.xlabel('Features')
plt.ylabel('Correlation')
plt.title('Correlation between Selected Features and Target Variable')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

# create a random forest classifier
clf = RandomForestClassifier(n_estimators=100, random_state=0)

# fit the model
clf.fit(X, y)

# get feature importances
feature_importances = clf.feature_importances_
# create a data frame for visualization.
final_df = pd.DataFrame({"Features": pd.DataFrame(X).columns, "Importances":feature_importances})
final_df.set_index('Importances')

# sort in descending order for better visualization
final_df = final_df.sort_values('Importances',ascending=False)

# Plotting the feature importances
plt.figure(figsize=(10, 6))
plt.bar(final_df['Features'][0:10], final_df['Importances'][0:10], color='teal')
plt.xlabel('Features')
plt.ylabel('Importances')
plt.title('Top 10 Feature Importances')
plt.xticks(rotation=90)
plt.show()

```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
def scale_features(df, scaler):
    # Scale only columns that aren't 'Category'
    to_scale = [col for col in df.columns if col != 'Category']
    scaled_features = scaler.fit_transform(df[to_scale])
    df_scaled = pd.DataFrame(scaled_features, columns=to_scale, index=df.index)
    return pd.concat([df['Category'], df_scaled], axis=1)

scalers = [StandardScaler(), MinMaxScaler()]

features_selected = final_df['Features'].head(50)
features_selected
X = X[features_selected]

for scaler in scalers:
    X = scale_features(X, scaler)
X
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import RandomizedSearchCV
from tensorflow import keras
from tensorflow.keras.utils import to_categorical

def create_model():
    model = keras.Sequential([
        keras.layers.Dense(256, activation='sigmoid'),
        keras.layers.Dense(128, activation='sigmoid'),
        keras.layers.Dense(64, activation='sigmoid'),
        keras.layers.Dense(2, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Convert labels to categorical one-hot encoding
y_binary = to_categorical(y)

# wrap the model using the function you created
clf = KerasClassifier(build_fn=create_model, verbose=0)

# define the random search parameters
param_dist = {'batch_size': [10, 20, 40, 60, 80, 100],
              'epochs': [10, 50, 100]}

# Create Random Search
random = RandomizedSearchCV(estimator=clf, param_distributions=param_dist, n_iter=10, n_jobs=-1, cv=2)
random_result = random.fit(X, y_binary)

# Report Results
print(f"Best: {random_result.best_score_} using {random_result.best_params_}")
means = random_result.cv_results_['mean_test_score']
stds = random_result.cv_results_['std_test_score']
params = random_result.cv_results_['params']
for mean, stdev, param in zip(means, stdds, params):
    print(f"Means: {mean}, Stdev: {stdev} with: {param}")
```

```python
import tensorflow as tf
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, confusion_matrix
from numpy import argmax, mean

# Setup data and model
num_classes = 2
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(256, activation='sigmoid'),
    tf.keras.layers.Dense(128, activation='sigmoid'),
    tf.keras.layers.Dense(64, activation='sigmoid'),
    tf.keras.layers.Dense(num_classes, activation='softmax')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Define early stopping
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

accuracies = []
for train, test in kfold.split(X, y):
    y_train = tf.keras.utils.to_categorical(y.iloc[train], num_classes)
    y_test = tf.keras.utils.to_categorical(y.iloc[test], num_classes)

    # Train model with early stopping
    model.fit(X.iloc[train], y_train, batch_size=20, epochs=100, verbose=1, callbacks=[early_stopping], validation_data=(X.iloc[test], y_test))

    # Make predictions
    y_pred = argmax(model.predict(X.iloc[test]), axis=-1)

    # Score model
    score = accuracy_score(argmax(y_test, axis=-1), y_pred)
    accuracies.append(score)

    print(f"Fold accuracy: {score*100:.2f}%")
    print(confusion_matrix(argmax(y_test, axis=-1), y_pred))

# Print mean accuracy
print(f"Mean accuracy: {mean(accuracies)*100:.2f}%")
```
